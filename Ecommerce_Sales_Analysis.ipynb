```python
import pandas as pd
import random

random.seed(42)

# Define data parameters
regions = ['North', 'South', 'East', 'West']
categories = ['Electronics', 'Clothing', 'Home & Garden', 'Sports', 'Books']
salespersons = ['Alice', 'Bob', 'Carol', 'David', 'Emma', 'Frank']

# Generate 200 sales transactions
data = {
    'transaction_id': range(1001, 1201),
    'region': [random.choice(regions) for _ in range(200)],
    'category': [random.choice(categories) for _ in range(200)],
    'salesperson': [random.choice(salespersons) for _ in range(200)],
    'sales_amount': [round(random.uniform(50, 5000), 2) for _ in range(200)],
    'customer_id': [random.randint(5000, 5100) for _ in range(200)]
}

df = pd.DataFrame(data)
print(df.head(10))
print(f"\nDataset shape: {df.shape}")
```

```python
# 1. Total sales amount for each region
region_sales = df.groupby('region')['sales_amount'].sum().reset_index()

# 2. Number of transactions for each category
category_transactions = df.groupby('category')['sales_amount'].count().reset_index(name='transaction_count')

# 3. Average sales amount per salesperson
salesperson_avg = df.groupby('salesperson')['sales_amount'].mean().reset_index()

# 4. Sort regions by total sales descending
region_sales_sorted = region_sales.sort_values(by='sales_amount', ascending=False)

print(region_sales_sorted)
print(category_transactions)
print(salesperson_avg)
```

```python
# 1. Group by region AND category
region_category_sales = df.groupby(['region', 'category'])['sales_amount'].sum().reset_index()

# 2. Multiple aggregations for each salesperson
salesperson_stats = df.groupby('salesperson')['sales_amount'].agg(['sum', 'mean', 'count']).reset_index()

# 3. Top performer (highest total sales)
top_performer = salesperson_stats.sort_values(by='sum', ascending=False).iloc[0]

# 4. Category with maximum total revenue
category_sales = df.groupby('category')['sales_amount'].sum()
top_category = category_sales.idxmax()

print(region_category_sales)
print(salesperson_stats)
print("Top Performer:\n", top_performer)
print("Top Category:", top_category)
```

```python
def sales_range(x):
    return x.max() - x.min()
```

```python
region_detailed = df.groupby('region')['sales_amount'].agg(
    total_sales='sum',
    avg_sales='mean',
    min_sales='min',
    max_sales='max',
    sales_range=sales_range
).reset_index()

print(region_detailed)
```

```python
final_report = df.groupby('region').agg({
    'sales_amount': ['sum', 'mean'],
    'customer_id': 'count'
}).reset_index()

# Flatten multi-level column names
final_report.columns = ['region', 'total_sales', 'avg_sales', 'total_transactions']

print(final_report)
```

Single aggregation --> Produces simple columns

Dictionary-based aggregation  -> Creates multi-level column headers

Why flatten columns?

Easier to read
